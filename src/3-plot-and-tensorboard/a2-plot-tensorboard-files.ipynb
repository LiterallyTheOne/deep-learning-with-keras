# <div style="text-align: center; color: cyan">Plot TensorBoard files</div>

## <div style="text-align: center; color: lime">Imports</div>

```python
from tbparse import SummaryReader
import matplotlib.pyplot as plt

from pathlib import Path
```

## <div style="text-align: center; color: lime">Load Tensorboard files</div>

```python
def load_tb_log(log_dir: Path):
    reader = SummaryReader(log_dir)
    df = reader.tensors

    loss = df[df["tag"] == "epoch_loss"]
    accuracy = df[df["tag"] == "epoch_accuracy"]

    return loss, accuracy
```

```python
e1_train_loss, e1_train_accuracy = load_tb_log(Path("logs/fit/20251120-101109/train"))
e1_val_loss, e1_val_accuracy = load_tb_log(Path("logs/fit/20251120-101109/validation"))

e2_train_loss, e2_train_accuracy = load_tb_log(Path("logs/fit/20251120-171107/train"))
e2_val_loss, e2_val_accuracy = load_tb_log(Path("logs/fit/20251120-171107/validation"))
```

```python
plt.figure()
plt.title("loss")
plt.plot(e1_train_loss["step"], e1_train_loss["value"], label="e1 train")
plt.plot(e1_val_loss["step"], e1_val_loss["value"], label="e1 validation")
plt.plot(e2_train_loss["step"], e2_train_loss["value"], label="e2 train")
plt.plot(e2_val_loss["step"], e2_val_loss["value"], label="e2 validation")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.title("accuracy")
plt.plot(e1_train_accuracy["step"], e1_train_accuracy["value"], label="e1 train")
plt.plot(e1_val_accuracy["step"], e1_val_accuracy["value"], label="e1 validation")
plt.plot(e2_train_accuracy["step"], e2_train_accuracy["value"], label="e2 train")
plt.plot(e2_val_accuracy["step"], e2_val_accuracy["value"], label="e2 validation")
plt.legend()
plt.show()
```
