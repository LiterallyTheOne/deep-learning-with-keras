## <div style="text-align: center; color: lime">Setup</div>

```python
import os

os.environ["KERAS_BACKEND"] = "torch"
```

## <div style="text-align: center; color: lime">Imports</div>

```python
from keras import Sequential
from keras.layers import Input, Conv2D, MaxPool2D, Dense, Flatten, Dropout

import numpy as np
```

## <div style="text-align: center; color: lime">Part 1</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 2</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 3</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
        Conv2D(256, (5, 5), padding="same", activation="relu"),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 4</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
        Conv2D(256, (5, 5), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 5</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
        Conv2D(256, (5, 5), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 6</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
        Conv2D(256, (5, 5), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 7</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
        Conv2D(256, (5, 5), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
        Conv2D(256, (3, 3), padding="same", activation="relu"),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 8</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
        Conv2D(256, (5, 5), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
        Conv2D(256, (3, 3), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 9</div>

```python
model = Sequential(
    [
        Input((224, 224, 3)),
        Conv2D(
            96, (11, 11), strides=4, padding="valid", activation="relu"
        ),
        MaxPool2D((3, 3), strides=2),
        Conv2D(256, (5, 5), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
        Conv2D(384, (3, 3), padding="same", activation="relu"),
        Conv2D(256, (3, 3), padding="same", activation="relu"),
        MaxPool2D((3, 3), strides=2),
        Flatten(),
        Dense(4096, activation="relu"),
        Dropout(0.5),
        Dense(4096, activation="relu"),
        Dropout(0.5),
        Dense(1000, activation="softmax"),
    ]
)

image = np.random.random((1, 224, 224, 3))

model(image).shape
```
