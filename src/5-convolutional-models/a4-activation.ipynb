## <div style="text-align: center; color: lime">Imports</div>

```python
from matplotlib import pyplot as plt

import torch

from keras.activations import tanh, relu, sigmoid, softmax
```

## <div style="text-align: center; color: lime">ReLU</div>

```python
x = torch.linspace(-10, 10, 1000)
y = relu(x).cpu()

plt.plot(x, y)

plt.show()
```

```python
a = 2 * torch.rand((1, 4, 4, 2)) - 1

result_activation = relu(a).cpu()

print("difference in shapes")
print(a.shape)
print(result_activation.shape)
print("-" * 20)

print("result of the 1st channel")
print(result_activation[0, :, :, 0])
print("-" * 20)

print("result of the 2nd channel")
print(result_activation[0, :, :, 1])
print("-" * 20)

print(f"result_max: {result_activation.numpy().max()}")
print(f"result_min: {result_activation.numpy().min()}")
```
