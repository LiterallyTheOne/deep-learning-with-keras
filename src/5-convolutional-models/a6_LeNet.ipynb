## <div style="text-align: center; color: lime">Setup</div>

```python
import os

os.environ["KERAS_BACKEND"] = "torch"
```

## <div style="text-align: center; color: lime">Imports</div>

```python
import keras

from keras import Sequential
from keras.layers import Input, Conv2D, AveragePooling2D, Dense, Flatten

import numpy as np
```

## <div style="text-align: center; color: lime">Part 1</div>

```python
model = Sequential(
    [
        Input((28, 28, 1)),
        Conv2D(6, (5, 5), padding="same", activation="sigmoid"),
    ]
)

image = np.random.random((1, 28, 28, 1))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 2</div>

```python
model = Sequential(
    [
        Input((28, 28, 1)),
        Conv2D(6, (5, 5), padding="same", activation="sigmoid"),
        AveragePooling2D((2, 2), strides=2),
    ]
)

image = np.random.random((1, 28, 28, 1))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 3</div>

```python
model = Sequential(
    [
        Input((28, 28, 1)),
        Conv2D(6, (5, 5), padding="same", activation="sigmoid"),
        AveragePooling2D((2, 2), strides=2),
        Conv2D(16, (5, 5), padding="valid", activation="sigmoid"),
    ]
)

image = np.random.random((1, 28, 28, 1))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 4</div>

```python
model = Sequential(
    [
        Input((28, 28, 1)),
        Conv2D(6, (5, 5), padding="same", activation="sigmoid"),
        AveragePooling2D((2, 2), strides=2),
        Conv2D(16, (5, 5), padding="valid", activation="sigmoid"),
        AveragePooling2D((2, 2), strides=2),
    ]
)

image = np.random.random((1, 28, 28, 1))

model(image).shape
```

## <div style="text-align: center; color: lime">Part 5</div>

```python
model = Sequential(
    [
        Input((28, 28, 1)),
        Conv2D(6, (5, 5), padding="same", activation="sigmoid"),
        AveragePooling2D((2, 2), strides=2),
        Conv2D(16, (5, 5), padding="valid", activation="sigmoid"),
        AveragePooling2D((2, 2), strides=2),
        Flatten(),
        Dense(120, activation="sigmoid"),
        Dense(84, activation="sigmoid"),
        Dense(10, activation="softmax"),
    ]
)

image = np.random.random((1, 28, 28, 1))

model(image).shape
```
