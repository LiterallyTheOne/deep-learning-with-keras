## <div style="text-align: center; color: lime">Imports</div>

```python
from pathlib import Path

from matplotlib import pyplot as plt

import torch
from torch.utils.data import random_split, DataLoader

from torchvision.datasets import ImageFolder
from torchvision import transforms

import kagglehub
```

## <div style="text-align: center; color: lime">Load the data</div>

```python
path = kagglehub.dataset_download("balabaskar/tom-and-jerry-image-classification")

data_path = Path(path) / "tom_and_jerry/tom_and_jerry"

all_data = ImageFolder(data_path)

g1 = torch.Generator().manual_seed(20)
train_data, val_data, test_data = random_split(all_data, [0.7, 0.2, 0.1], g1)
```

## <div style="text-align: center; color: lime">Plot 10 data of each class</div>

```python
NUMBER_OF_SAMPLES = 10
```

```python
num_classes = len(all_data.classes)
print("number of classes: ", num_classes)
```

```python
image_samples = {}
for i in range(num_classes):
    image_samples[i] = []
```

```python
for image, label in train_data:
    if len(image_samples[label]) < NUMBER_OF_SAMPLES:
        image_samples[label].append(image)
    all_lens = [len(x) for x in image_samples.values()]
    if sum(all_lens) >= NUMBER_OF_SAMPLES * num_classes:
        break
```

```python
fig, axes = plt.subplots(NUMBER_OF_SAMPLES, num_classes, figsize=(16, 16))

for i in range(NUMBER_OF_SAMPLES):
    for j in range(num_classes):
        axes[i, j].set_axis_off()
        axes[i, j].imshow(image_samples[j][i])
```
