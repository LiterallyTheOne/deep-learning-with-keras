# <div style="text-align: center; color: cyan">Dropout layer</div>

## <div style="text-align: center; color: lime">Setup</div>

```python
import os

os.environ["KERAS_BACKEND"] = "torch"
```

## <div style="text-align: center; color: lime">Imports</div>

```python
from keras import Sequential
from keras.layers import Dropout, Dense, Input
```

## <div style="text-align: center; color: lime">Dropout layer</div>

```python
model = Sequential(
    [
        Input(shape=(1024,)),
        Dropout(0.2),
        Dense(512, activation='relu'),
        Dropout(0.2),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='relu'),
    ]
)
```
