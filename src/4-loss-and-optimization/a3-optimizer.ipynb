## <div style="text-align: center; color: lime">SGD</div>

Simple SGD Formula:

$$
\theta_{t+1} = \theta_t - \eta \, \nabla_{\theta} \, L(\theta_t; x_{t}, y_{t})
$$

* pros:
    * Memory efficient
    * Good Generalization
    * Simple
    * Good for smooth problems (convex)
* cons:
    * Slow convergence
    * Sensitive to Learning rate
    * Can get stuck

```python
from keras.optimizers import SGD

sgd = SGD()
```

## <div style="text-align: center; color: lime">RMSProp</div>

* pros:
    * Adaptive Learning rate
    * Super good for RNNs
    * low-memory
    * More stable than SGD
* cons:
    * Sensitive to hyperparameters
    * Not as good at Generalization as SGD

```python
from keras.optimizers import RMSprop

rmsprop = RMSprop()
```

## <div style="text-align: center; color: lime">Adam</div>

* pros:
    * Fast and stable convergence
    * Handles sparse gradients pretty well
    * Good for non-stationary problems
        * Reinforcement Learning
        * GAN
    * Has bias correction
        * which improves training
* cons:
    * Sometimes not having a better generalization than SGD
    * Uses more memory than SGD and RMSprop

```python
from keras.optimizers import Adam

adam = Adam()
```
